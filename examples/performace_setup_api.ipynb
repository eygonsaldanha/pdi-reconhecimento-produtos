```python
import multiprocessing
import time
from datetime import datetime

def print_timestamp(i):
    """Função que imprime o índice e o timestamp atual."""
    print(f"[{i:03}] {datetime.now().strftime('%Y-%m-%d %H:%M:%S.%f')}")
    time.sleep(0.01)
    
def main():
    total_prints = 500
    num_processes = 5

    with multiprocessing.Pool(processes=num_processes) as pool:
        pool.map(print_timestamp, range(total_prints))

if __name__ == "__main__":
    main()
```
