```python
import cv2
import numpy as np
```

```python
image_path = '../../dataset/Limes/113_100.jpg'
img_to_test = cv2.imread(image_path)
img_to_test
```

```python
from preprocessing import converter_para_cinza
img_to_test_cinza = converter_para_cinza(img_to_test)
img_to_test_cinza
```

```python
from preprocessing import aplicar_filtro_gaussiano
img_to_test_gaussiano = aplicar_filtro_gaussiano(img_to_test_cinza)
img_to_test_gaussiano
```

```python
from segmentation import segmentar_objeto_com_flood_fill
img_to_test_mascara = segmentar_objeto_com_flood_fill(img_to_test_gaussiano)
img_to_test_mascara
```

```python
from segmentation import encontrar_contornos
img_to_test_contornos = encontrar_contornos(img_to_test_mascara)
img_to_test_contornos
```

```python
from geometric import calcular_area, calcular_perimetro, calcular_circularidade, calcular_aspect_ratio

for i, contorno in enumerate(img_to_test_contornos):
    area = calcular_area(contorno)
    perimetro = calcular_perimetro(contorno)
    circularidade = calcular_circularidade(contorno)
    aspect = calcular_aspect_ratio(contorno)
    
    print(f"\nContorno {i+1}:")
    print(f"  Área: {area:.2f} pixels")
    print(f"  Perímetro: {perimetro:.2f}")
    print(f"  Circularidade: {circularidade:.4f}")
    print(f"  Aspect Ratio: {aspect:.4f}")
```
